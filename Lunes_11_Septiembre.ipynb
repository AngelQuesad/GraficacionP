```python
#Apertura de archivo
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos=pd.read_csv('/content/casasboston.csv')
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df = datos.rename( columns= {
    "TOWN":"CIUDAD",
    "CRIM":"CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO CHARLES",
    "RM":"N_HABITANTES_MEDIO",
    "MEDV":"VALOR_MEDIO",
    "LSTAT":"PCT_CLASE_BAJA"})

print(df)
```

```python
df.RM.plot.hist()
plt.show()
```

```python
#Gráfico actualizado
fig, ax = plt.subplots(figsize=(2, 2), facecolor='lightgreen',
                       layout='constrained')
fig.suptitle('Figura')
ax.set_title('HABITANTES MEDIOS', loc='right', fontstyle='oblique', fontsize='large')
df.N_HABITANTES_MEDIO.plot.hist()
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIO",y="CRIMEN",alpha=0.9)
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIO"].mean()
```

```python
valor_por_ciudad.head(10).plot.barh()
plt.show()
```
